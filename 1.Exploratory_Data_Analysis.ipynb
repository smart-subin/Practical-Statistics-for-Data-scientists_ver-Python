```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
root = os.path.abspath("./MyDrive/data-stduy/")
```

```python
csv = os.path.join(root, "1.csv")
data = pd.read_csv(csv)
data
```

```python
def get_avg(datas):
  return sum(datas) / len(datas)

def get_cut_avg(datas, trim:int=1):
  datas = list(datas)
  datas.sort()
  return sum(datas[trim: -1*trim]) / (len(datas) - (trim * 2))

population = data["인구"]
avg = get_avg(population)
cut_avg = get_cut_avg(population)
```

```python
def get_weighted_avg(datas, w):
  return sum(datas * w) / sum(w)

def get_cut_weighted_avg(datas, trim:int=1):
  datas.sort(key=lambda x: x[0] * x[1])
  return sum(list(map(lambda x: x[0] * x[1],datas[trim: -1*trim]))) / sum(list(map(lambda x: x[1],datas[trim: -1*trim])))

murder_rate = data["살인 비율"]
population = data["인구"]
get_weighted_avg(murder_rate, population)
get_cut_weighted_avg(data[["살인 비율", "인구"]].values.tolist(), 1)
```
